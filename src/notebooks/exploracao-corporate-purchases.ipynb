```python
# Convert `created` column to datetime
df['created'] = pd.to_datetime(df['created'], format="%Y-%m-%d %H:%M:%S.%f UTC")

# Summary statistics for numeric columns
print("\nSummary Statistics for Numeric Columns:\n")
print(df.describe().to_markdown(numalign="left", stralign="left"))

# Value counts for categorical columns
categorical_cols = ['status', 'merchantName', 'merchantCategoryType']

for col in categorical_cols:
    print(f"\nValue Counts for {col}:\n")
    print(df[col].value_counts().reset_index().rename(columns={'index':col, col:'count'}).to_markdown(index=False, numalign="left", stralign="left"))
```
